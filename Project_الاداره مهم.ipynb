```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/dataset.csv')

# Fill missing values for categorical columns with 'Unknown'
df.fillna({col: 'Unknown' for col in df.select_dtypes(include='object').columns}, inplace=True)

# Remove duplicate rows
df = df.drop_duplicates()

# Drop a column if it's not needed
df = df.drop(columns=['Networking Importance'], errors='ignore')
df = df.drop(columns=['Undergrad University Ranking'], errors='ignore')
df = df.drop(columns=['Location Preference (Post-MBA)'], errors='ignore')

# Save the Cleaned Data
df.to_csv('cleaned_dataset.csv', index=False)

print("Data cleaning completed! Cleaned dataset saved as 'cleaned_dataset.csv'.")
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Load the dataset
data = pd.read_csv('/content/cleaned_dataset.csv')

# Check data types to identify numerical and categorical columns
print("\nDataset Information:")
print(data.info())

# Ensure numerical and categorical columns are properly defined
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = data.select_dtypes(include=['object']).columns

print("\nNumerical Columns:", numerical_columns.tolist())
print("Categorical Columns:", categorical_columns.tolist())

```

```python
# -------- Univariate Analysis --------

# Analyze central tendencies and spread for numerical features
if len(numerical_columns) > 0:
    print("\nCentral Tendencies and Spread for Numerical Features:")
    for column in numerical_columns:
        print(f"\n{column}:")
        print(f"  Mean: {data[column].mean():.2f}")
        print(f"  Median: {data[column].median():.2f}")
        print(f"  Mode: {data[column].mode().values[0]}")
        print(f"  Variance: {data[column].var():.2f}")
        print(f"  Standard Deviation: {data[column].std():.2f}")
        print(f"  Min: {data[column].min()}")
        print(f"  Max: {data[column].max()}")
else:
    print("\nNo numerical columns to analyze.")

# Visualize the distribution of numerical features
if len(numerical_columns) > 0:
    print("\nVisualizing Numerical Features:")
    for column in numerical_columns:
        plt.figure(figsize=(8, 5))
        sns.histplot(data[column], kde=True, bins=30)
        plt.title(f"Distribution of {column}")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.show()
else:
    print("\nNo numerical columns to visualize.")

# Visualize the count of categorical features
if len(categorical_columns) > 0:
    print("\nVisualizing Categorical Features:")
    for column in categorical_columns:
        plt.figure(figsize=(8, 5))
        sns.countplot(y=data[column], order=data[column].value_counts().index)
        plt.title(f"Count of {column}")
        plt.xlabel("Count")
        plt.ylabel(column)
        plt.show()
else:
    print("\nNo categorical columns to visualize.")
```
